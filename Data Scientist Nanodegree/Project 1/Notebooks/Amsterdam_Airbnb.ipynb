## 1) Business Understanding

Amsterdam is a popular tourist destination in the Netherlands. With its picturesque canals and bridges, historic homes and sights including the Anne Frank house and Van Gogh museum, Amsterdam attracts a lot of tourists through out the year. This makes holiday rental homes like Airbnbs a lucrative option for the locals to earn an additional income. 

However, there has been claims that the popularity of Airbnbs in Amsterdam has led to gentrification, thereby driving up real-estate prices and increasing the risk of a housing market bubble [3, 4]. This also has created a shortage in housing for locals, especially in popular tourist place like the Center of Amsterdam [5]. 

To tackle these problems, the Municipality of Amstedam has taken some stringent measures including a permit system, that limits renting out homes to tourists for a maximum of 30 nights per year to a group of atmost 4 guests [6, 7]. Short-term rentals in Amsterdam must either be classified as a holiday rental or a bed & breakfast (B&Bs), with different regulations applying for each of these two categories. Non-compliance to the regulations stipulated by the Municipality can lead to fines upto the tune of 9000 euros [8]. 

The regulations for holiday rentals include the following [8,9]:

    1) The property must be the main residence of the host. 
    2) Only the main resident may rent their entire primary home to others, and cannot be present during the duration of the rental.
    3) One can rent for a maximum of 30 nights per year.
    4) One can only rent to a maximum of four people at once.


The regulations for B&Bs (renting out private room fall under this category) include the following [8,9]:

    1) A maximum of 40% of your residence may be rented out.
    2) You may not rent an independent living space (for example with a private kitchen or separate entrance).
    3) One can only rent to a maximum of four people at once.
    4) Only the main resident can rent the space, and must be present during the rental.
    5) There are no limit on the number of nights that a listing can be rented out and can be rented the entire year. 
    
Such regulations have made it difficult for new hosts to break into the short-rental business in Amsterdam. In this project, we will try to answer a few questions that will help new hosts in making informed decisions about renting their property in Amsterdam. 

We will try to answer the following questions:

    1) Which neighbourhood is the most popular among tourists in Amsterdam? What are the average prices in the neighbourhoods? Is there a trend in the prices and reviews for the neighbourhoods? 
    2) What are the type of listings in Amsterdam and what is their average price? What are the most sought-after amenities provided by hosts in Amsterdam?
    3) What time of the year is the demand high for listings in Amsterdam? How does the price of a listing change through the year? 
    4) What factors affect the price of a listing?
    5) How effective are the regulations imposed by the municipality? Are there any listings that are breaking the rules?

        > - **`date`** will have to be converted to `datetime`

```python
# A function to convert datatypes

def convert_datatypes(df, cols_date=[], cols_perc=[], cols_bool=[], cols_curr=[]):
    '''
    Function to convert date columns to datetime format
    
    Arguments:
    df - a dataframe
    cols_date - a list of columns to convert to date
    cols_perc - a list of columns to convert to percentage (float)
    cols_bool - a list of columns to convert to bool
    cols_curr - a list of columns to convert to float
    
    '''
    for col in cols_date:
        if df[col].dtype == 'object':
                try:
                    df[col] = pd.to_datetime(df[col])
                except ValueError:
                    pass
                
    for col in cols_perc:
        if df[col].dtype == 'object':
                try:
                    df[col] = df[col].replace('%', '', regex=True).astype(float)
                    df[col] = df[col] * 0.01
                except ValueError:
                    pass
                
    for col in cols_bool:
        if df[col].dtype == 'object':
                try:
                    df[col] = df[col].apply(lambda x: 1 if x == 't' else 0)
                except ValueError:
                    pass
                
    for col in cols_curr:
        if df[col].dtype == 'object':
                try:
                    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)
                except ValueError:
                    pass
```

```python
def fit_linear_mod(df, response_col, drop_cols, test_size=.3, rand_state=42):
    
    
    '''
    A function to fit a linear model after some cleaning of the set of features
    
    Arguments:
    df - a dataframe holding all the columns
    response_col - a string holding the name of response variable
    drops_cols - columns to drop for creating the feature set
    test_size - a float between [0,1] about what proportion of data should be in the test dataset
    rand_state - an int that is provided as the random state for splitting the data into training and test
    
    Returnss:
    X - cleaned X matrix 
    y - cleaned response variable
    test_score - float - r2 score on the test data
    train_score - float - r2 score on the test data
    lm_model - model object from sklearn
    X_train, X_test, y_train, y_test - output from sklearn train test split used for optimal model
    '''

    # Drop the columns that are not necessary for the analysis
    df = df.drop(drop_cols, axis=1)
    
    #Pull a list of the column names of the categorical variables
    cat_df = df.select_dtypes(include=['object'])
    cat_cols = cat_df.columns
    
    #dummy all the cat_cols
    for col in  cat_cols:
        df = pd.concat([df.drop(col, axis=1), pd.get_dummies(df[col], prefix=col, prefix_sep='_', drop_first=True, dummy_na=True)], axis=1)

    #Split into explanatory and response variables
    X = df.drop(response_col, axis=1)
    y = df[response_col]
    
    # Remove overfitting columns    
    overfit = []
    for i in X.columns:
        counts = X[i].value_counts()
        zeros = counts.iloc[0]
        if zeros / len(X) * 100 > 95:
            overfit.append(i)
    overfit
    
    X = X.drop(overfit, axis=1)

    #Split into train and test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=rand_state)

    lm_model = LinearRegression() # Instantiate
    lm_model.fit(X_train, y_train) #Fit

    #Predict using your model
    y_test_preds = lm_model.predict(X_test)
    y_train_preds = lm_model.predict(X_train)

    #Score using your model
    test_score = r2_score(y_test, y_test_preds)
    train_score = r2_score(y_train, y_train_preds)

    return X, y, test_score, train_score, lm_model, X_train, X_test, y_train, y_test, y_test_preds, y_train_preds
```

```python
X, y, test_score, train_score, lm_model, X_train, X_test, y_train, y_test, y_test_preds, y_train_preds = fit_linear_mod(df, response_col, drop_cols, test_size=.3, rand_state=42)
```

```python
def coef_weights(lm_model, X_train):
    
    '''
    Function of get the coefficients from the linear model object
    
    Arguments:
    lm_model - object of the linear model 
    X_train - the training data, so the column names can be used
    Returnds:
    coefs_df - a dataframe holding the coefficient and estimates
    '''
    
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = X_train.columns
    coefs_df['coefs'] = lm_model.coef_
    return coefs_df

coef_df = coef_weights(lm_model, X_train)
```

```python
# Features that have a negative influence on the price

coef_df[coef_df['coefs']<0].sort_values(by='coefs', ascending=False)
```
